```python
# Assignment 9
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
# using the zip function the *matrix separates the row and the zip funciton pairs elements column wise
transpose = list(zip(*matrix))
print(transpose)
```

```python
# Assignment 14
lst = list(range(1,21))
print(lst)
n = 4
# right rotate
result1 = lst[-n:] + lst[:len(lst)-n]

# left rotate
result2 = lst[n:] + lst[:n]

print(result1)
print(result2)
```

```python
# Assignment 15
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

intersection = [x for x in a if x in b]
print(intersection)
```
